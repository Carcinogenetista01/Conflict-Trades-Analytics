```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('../Inflacion/inflacion_imf_clean.csv')
df.set_index('Country', inplace=True)
df.columns = df.columns.astype(str) 
```

```python
inflacion_promedio = df.loc[:, '2000':'2024'].mean()
inflacion_promedio.plot(figsize=(10, 5), title='Inflación promedio global (2000–2024)')
plt.xlabel('Año')
plt.ylabel('Inflación (%)')
plt.grid(True)
plt.show()
```

```python
inflacion_2023 = df['2023'].sort_values(ascending=False).head(5)
inflacion_2023.plot(kind='bar', title='Top 5 países con mayor inflación en 2023')
plt.ylabel('Inflación (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
ultimos_10_años = [str(año) for año in range(2014, 2024)]
inflacion_baja = df[ultimos_10_años].mean(axis=1)
inflacion_estable = inflacion_baja[inflacion_baja < 3].sort_values().head(5)
inflacion_estable.plot(kind='barh', title='Top 5 países con inflación baja y estable (2014–2023)')
plt.xlabel('Inflación promedio (%)')
plt.grid(axis='x')
plt.show()
```

```python
paises = ['Australia', 'Austria', 'Albania', 'Algeria', 'Andorra']
valores = df.loc[paises, '2023']
plt.figure(figsize=(6, 6))
plt.pie(valores, labels=paises, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de inflación entre 5 países (2023)')
plt.show()
```

```python
subset = df.loc[:, '2020':'2024'].sort_values('2024', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.heatmap(subset, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Heatmap de inflación (2020–2024) – Top 10 países')
plt.ylabel('País')
plt.xlabel('Año')
plt.show()
```

```python
df = pd.read_csv('../Life_expectancy/life_expectancy_un_clean.csv')
df.set_index('Country', inplace=True)
df.columns = df.columns.astype(str) 
```

```python

df = df.drop(columns=["ISO3"]) 
df.columns = df.columns.astype(str)
promedio_global = df.mean(axis=0)

plt.figure(figsize=(10, 5))
plt.plot(promedio_global, marker='o')
plt.title("Esperanza de vida promedio global (1990–2025)")
plt.xlabel("Año")
plt.ylabel("Esperanza de vida (años)")
plt.grid(True)

ticks = [str(año) for año in range(1990, 2030, 10)]
plt.xticks(ticks)

plt.show()
```

```python
vida_2025 = df["2025"].sort_values(ascending=False).head(5)

vida_2025.plot(kind="bar", color="mediumseagreen", title="Top 5 países con mayor esperanza de vida (2025)")
plt.ylabel("Esperanza de vida (años)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
vida_1990 = df["1990"].sort_values().head(5)

vida_1990.plot(kind="barh", color="tomato", title="Top 5 países con menor esperanza de vida (1990)")
plt.xlabel("Esperanza de vida (años)")
plt.grid(axis="x")
plt.show()
```

```python
paises = ["Afghanistan", "Albania", "Algeria", "Andorra", "American Samoa"]
valores = df.loc[paises, "2023"]

plt.figure(figsize=(6, 6))
plt.pie(valores, labels=paises, autopct='%1.1f%%', startangle=90)
plt.title("Distribución de esperanza de vida (2023)")
plt.show()
```

```python
import seaborn as sns

paises_clave = ["Afghanistan", "Albania", "Algeria", "Andorra", "American Samoa"]
años_recientes = [str(año) for año in range(2010, 2026)]

heatmap_df = df.loc[paises_clave, años_recientes]

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_df, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Evolución de esperanza de vida (2000–2025)")
plt.xlabel("Año")
plt.ylabel("País")
plt.show()
```
